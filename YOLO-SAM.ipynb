```python
from ultralytics import YOLO

model = YOLO("yolo11n.pt")  # or "yolov8n.pt" depending on version
results = model("image.jpg")  # your test image

results[0].plot()            # visualize detections
boxes = results[0].boxes.xyxy.cpu().numpy()  # bounding boxes [x1,y1,x2,y2]
```

```python
import requests
import os

url = "https://dl.fbaipublicfiles.com/segment_anything/sam_vit_b_01ec64.pth"
print("Downloading 375MB SAM checkpoint...")
r = requests.get(url, stream=True)
with open("sam_vit_b.pth", "wb") as f:
    for chunk in r.iter_content(8192):
        f.write(chunk)
print(f"✅ Complete: {os.path.getsize('sam_vit_b.pth')/1e6:.1f} MB")
```

```python

import os
size_mb = os.path.getsize("sam_vit_b.pth") / 1e6 if os.path.exists("sam_vit_b.pth") else 0
print(f"SAM checkpoint: {'✅ READY ({size_mb:.1f} MB)' if size_mb > 300 else '❌ MISSING'}")
```

```python

from segment_anything import sam_model_registry, SamPredictor
import cv2; import numpy as np

image = cv2.imread("image.jpg")
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

sam = sam_model_registry["vit_b"](checkpoint="sam_vit_b.pth")
print("✅ SAM LOADED SUCCESSFULLY!")
predictor = SamPredictor(sam)
predictor.set_image(image_rgb)
print("✅ Full YOLO+SAM pipeline ready!")
```

```python
from segment_anything import sam_model_registry, SamPredictor
import cv2; import numpy as np

image = cv2.imread("image.jpg")
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

sam = sam_model_registry["vit_b"](checkpoint="sam_vit_b.pth")  # Now works
predictor = SamPredictor(sam)
predictor.set_image(image_rgb)

box = boxes[0] * np.array([image_rgb.shape[1], image_rgb.shape[0], image_rgb.shape[1], image_rgb.shape[0]])
masks, scores, _ = predictor.predict(box=box.astype(int))
print(f"SAM success! Mask score: {scores[0]:.3f}")
```
